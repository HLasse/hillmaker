```python
opacity = 0.5
total_df['capacity'] = 170
time_bins = np.arange(336)
plt.bar(time_bins,total_df['mean'], color='green', alpha=opacity) 
plt.plot(time_bins,total_df['p95'], color='black')
plt.plot(time_bins,total_df['capacity'], color='blue')
plt.xlabel('Bin of week')
plt.ylabel('Occupancy')
plt.title(r'Total Occupancy')
plt.axis([0, 335, 0, 250])
plt.grid(True)
fig1 = matplotlib.pyplot.gcf()
fig1.set_size_inches(16,12)
plt.show()
```

Occupancy graph patterns
------------------------

Let's create some common patterns using matplotlib and ggplot.

```python
# Create a Figure and Axes object
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)

# Specify the X-axis values
time_bins = range(1,337)

# Specify the mean occupancy values and upper percentile values
mean_occ = total_df['mean']
pctile_occ = total_df['p95']

# Specify a capacity line
capacity = total_df['capacity']

# Style the bars for mean occupancy
bar_color = 'green'
bar_opacity = 0.8

# Style the line for the occupancy percentile and capacity
pctile_line_style = 'k--'
capacity_line_style = 'k-'

pctile_line_width = 1
capacity_line_width = 2

# Add data to the plot
ax1.bar(time_bins,mean_occ, color=bar_color, alpha=bar_opacity)
ax1.plot(time_bins,pctile_occ, linestyle=pctile_line_style)

# Set plot and axis titles
ax1.set_title('Total Occupancy')
ax1.set_xlabel('Time Bin of Week')
ax1.set_ylabel('Occupancy')

# Gridlines
ax1.grid(True, color='k')

# Legend

# Plot size
fig1.set_size_inches(16,12)
```
